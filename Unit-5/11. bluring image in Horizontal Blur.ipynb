```python
# 11. Bluring an image using inbulit function in Horizontal Blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('au.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

size = 30
k1 = np.zeros((size,size))
k1 [int(size/2),:] = np.ones(size)
k1 = k1/size

horizontal = cv2.filter2D(img,-1,k1)

plt.subplot(1,2,1),plt.imshow(img),plt.title("Source Image"),plt.axis("off")
plt.subplot(1,2,2),plt.imshow(horizontal),plt.title("Horizontal Image"),plt.axis("off")
```
