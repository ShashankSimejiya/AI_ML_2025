```python
# 18. Gussian Kernel to generate more natural enhancement

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('aerial.webp')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gauss = cv2.GaussianBlur(img,(5,5),0)
sharped_img = cv2.addWeighted(img,1.5,gauss,-0.5,0)

plt.subplot(1,2,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(sharped_img),plt.title('Sharped_Image'),plt.axis('off')
```
