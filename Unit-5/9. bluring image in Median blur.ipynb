```python
# 9. Bluring a image in Median Blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('au.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

mead = cv2.medianBlur(img,(19))

plt.subplot(1,2,1),plt.imshow(img),plt.title("Source Blur"),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(mead),plt.title("Median Blur"),plt.axis('off')
```
