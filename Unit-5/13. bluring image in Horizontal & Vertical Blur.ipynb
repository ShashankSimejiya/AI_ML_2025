```python
# 13. Bluring an image using inbulit function in Horizontal && Vertical Blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('au.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

# Horizontal Blur
size = 30
k1 = np.zeros((size,size))
k1 [int(size/2),:] = np.ones(size)
k1 = k1/size

# Vertical Blur
size = 30
k2 = np.zeros((size,size))
k2 [:,int(size/2)] = np.ones(size)
k2 = k2/size

horizontal = cv2.filter2D(img,-1,k1)
vertical = cv2.filter2D(img,-1,k2)


plt.subplot(1,3,1),plt.imshow(img),plt.title("Source Image"),plt.axis("off")
plt.subplot(1,3,2),plt.imshow(horizontal),plt.title("Horizontal Image"),plt.axis("off")
plt.subplot(1,3,3),plt.imshow(vertical),plt.title("Vertical Image"),plt.axis("off")
```
