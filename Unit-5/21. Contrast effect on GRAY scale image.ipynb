```python
# 21. Contrast Effect on GRAY Scale Image

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('flower.jpg',0)
contrast = cv2.equalizeHist(img)

plt.subplot(1,2,1),plt.imshow(img,cmap='gray'),plt.title('Source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(contrast,cmap='gray'),plt.title('Contrast Image'),plt.axis('off')
```
