```python
# 27. Applying advantages image filters
# 3.) Scharr (for advantages smoothing effect)

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('flower.jpg',0)

scharrX = cv2.Scharr(img,cv2.CV_64F,1,0)
scharrY = cv2.Scharr(img,cv2.CV_64F,0,1)

scharr_X = cv2.convertScaleAbs(scharrX)
scharr_Y = cv2.convertScaleAbs(scharrY)

scharr_Combine = cv2.addWeighted(scharr_X,0.10,scharr_Y,0.2,0)

plt.subplot(1,4,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(1,4,2),plt.imshow(scharr_X,cmap='gray'),plt.title('Scharr-X'),plt.axis('off')
plt.subplot(1,4,3),plt.imshow(scharr_Y,cmap='gray'),plt.title('Scharr-Y'),plt.axis('off')
plt.subplot(1,4,4),plt.imshow(scharr_Combine,cmap='gray'),plt.title('Scharr-Combine'),plt.axis('off')
```
