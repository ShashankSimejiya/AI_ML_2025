```python
# 10. Bluring a image in Box && Gaussian && Median Blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('au.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

box = cv2.blur(img,(20,20))
gaus = cv2.GaussianBlur(img,(29,29),5)
mead = cv2.medianBlur(img,(19))

plt.subplot(1,4,1),plt.imshow(img),plt.title("Source Image"),plt.axis('off')
plt.subplot(1,4,2),plt.imshow(box),plt.title("Box Blur"),plt.axis('off')
plt.subplot(1,4,3),plt.imshow(gaus),plt.title("Gaussian Blur"),plt.axis('off')
plt.subplot(1,4,4),plt.imshow(mead),plt.title("Median Blur"),plt.axis('off')
```
