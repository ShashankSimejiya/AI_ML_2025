```python
# 22. Contrast Effect on Color Image using CLAHE

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('flower.jpg')

ycrcb = cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
y,cr,cb = cv2.split(ycrcb)

clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
y_clahe = clahe.apply(y)
ycrcb_clahe = cv2.merge([y_clahe,cr,cb])
clahe_eq = cv2.cvtColor(ycrcb_clahe,cv2.COLOR_YCrCb2RGB)
source = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.subplot(1,2,1),plt.imshow(source),plt.title('Source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(clahe_eq),plt.title('Contrast Color Image using CLAHE'),plt.axis('off')
```
