```python
# 8. Bluring a image in Gaussian Blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('au.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

gaus = cv2.GaussianBlur(img,(29,29),5)

plt.subplot(1,2,1),plt.imshow(img),plt.title("Source Blur"),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(gaus),plt.title("Gaussian Blur"),plt.axis('off')
```
