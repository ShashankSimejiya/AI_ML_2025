```python
# 7. Bluring a image in box blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('au.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

box = cv2.blur(img,(20,20))

plt.subplot(1,2,1),plt.imshow(img),plt.title("Source Blur"),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(box),plt.title("Box Blur"),plt.axis('off')
```
