```python
# 12. Bluring an image using inbulit function in Vertical Blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('au.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

size = 30
k2 = np.zeros((size,size))
k2 [:,int(size/2)] = np.ones(size)
k2 = k2/size

vertical = cv2.filter2D(img,-1,k2)

plt.subplot(1,2,1),plt.imshow(img),plt.title("Source Image"),plt.axis("off")
plt.subplot(1,2,2),plt.imshow(vertical),plt.title("Vertical Image"),plt.axis("off")
```
