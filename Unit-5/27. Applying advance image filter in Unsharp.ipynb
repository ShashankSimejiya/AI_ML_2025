```python
# 26. Applying advantages image filters
# 2.) Unsharp (for advantages smoothing effect)

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('flower.jpg')
gauss = cv2.GaussianBlur(img,(9,9),0)
unsharp = cv2.addWeighted(img,1.5,gauss,-0.5,0)

plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('Source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(unsharp,cv2.COLOR_BGR2RGB)),plt.title('With Smoothing effect'),plt.axis('off')
```
